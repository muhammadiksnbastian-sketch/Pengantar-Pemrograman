```python
import pandas as pd                                     # Mengimpor library pandas untuk mengolah data (terutama tabel/Excel), disingkat sebagai 'pd'

from scipy.stats.mstats import gmean, hmean             # Mengimpor fungsi gmean (geometric mean) dan hmean (harmonic mean) dari scipy
import matplotlib.pyplot as plt                         # Mengimpor matplotlib untuk membuat grafik/visualisasi, disingkat sebagai 'plt'


my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')      # Membaca file Excel dan menyimpannya ke dalam variabel 'data_excel' dan sheet_name digunakan untuk memilih sheet tertentu dalam file Excel

a_mean = my_dataset.Zr.mean()       # Menghitung rata-rata aritmatika (mean biasa) dari kolom 'Zr' pada dataset
g_mean = gmean(my_dataset['Zr'])    # Menghitung rata-rata aritmatika (mean biasa) dari kolom 'Zr' pada dataset
h_mean = hmean(my_dataset['Zr'])    # Menghitung rata-rata harmonik dari kolom 'Zr' menggunakan fungsi hmean

print ('-------')                           # Menampilkan garis pemisah di output
print ('arithmetic mean')                   # Menampilkan teks keterangan
print ("{0:.1f} [ppm]".format(a_mean))      # Menampilkan nilai mean aritmatika dengan 1 angka di belakang koma
print ('-------')                           # Garis pemisah lagi

print ('geometric mean')                    # Menampilkan teks "geometric mean" sebagai judul output
print ("{0:.1f} [ppm]".format(g_mean))      # Menampilkan nilai rata-rata geometri dengan 1 angka di belakang koma dan satuan ppm
print ('-------')                           # Menampilkan garis pemisah agar output lebih rapi

print ('harmonic mean')                     # Menampilkan teks "harmonic mean" sebagai judul output
print ("{0:.1f} [ppm]".format(h_mean))      # Menampilkan nilai rata-rata harmonik dengan format 1 angka di belakang koma
print ('-------')                           # Garis pemisah lagi

fig, ax = plt.subplots()                                                                                     # Membuat area gambar (figure) dan sumbu (axes) untuk plotting grafik
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8)       # Data kolom Zr yang akan dibuat histogram
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2)                                     # Jumlah interval histogram ditentukan otomatis
ax.axvline(g_mean, color='orange',  label='Geometric mean', linewidth=2)                                     # Menampilkan dalam bentuk distribusi (probability density)
ax.axvline(h_mean, color='blue',  label='Harmonic mean', linewidth=2)                                       # Warna garis tepi batang (k = black)
ax.set_xlabel('Zr [ppm]')                                       # Memberi label pada sumbu X (nilai Zr dalam satuan ppm)
ax.set_ylabel('Probability density')                            # Memberi label pada sumbu Y (kepadatan distribusi)
ax.legend()                                                     # Menampilkan legenda agar tiap warna/garis bisa dikenali

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

hist, bin_edges = np.histogram(my_dataset['Zr'], bins= 20, density=True)            # Menghitung histogram dari data kolom 'Zr' dengan 20 bin dan dinormalisasi (density=True), 
                                                                                    # hasilnya berupa nilai histogram (hist) dan batas tiap bin (bin_edges)
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2            # Menentukan nilai modus (modal value) dengan mengambil bin yang memiliki frekuensi tertinggi,
                                                                                    # lalu menghitung titik tengah dari bin tersebut

print ('modal value: {0:.0f} [ppm]'.format(modal_value))            # Menampilkan nilai modus dalam format tanpa desimal dengan satuan ppm

fig, ax = plt.subplots()        # Membuat figure dan axes (area plot) menggunakan matplotlib
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.8)      # Membuat histogram dari data 'Zr' dengan 20 bin, dinormalisasi, 
                                                                                                        # garis tepi hitam (edgecolor='k'), label, dan transparansi 0.8

ax.axvline(modal_value, color="orange", label="Modal value", linewidth=2)   # Menambahkan garis vertikal pada nilai modus dengan warna oranye dan ketebalan garis 2
ax.set_xlabel('Zr [ppm]')                   # Memberi label pada sumbu x (konsentrasi Zr dalam ppm)
ax.set_ylabel('Probability density')        # Memberi label pada sumbu y (kepadatan probabilitas)
ax.legend()                                 # Menampilkan legenda pada grafik

'''
Output: modal value: 277 [ppm]
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

variance =  my_dataset['Zr'].var()
stddev =  my_dataset['Zr'].std()

print ('-------')
print ('Variance')
print("{0:.0f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins= 20, density = True, edgecolor='k', label='Measurements Hist') 
ax.axvline(my_dataset['Zr'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2)
ax.axvline(my_dataset['Zr'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2)
ax.axvspan(my_dataset['Zr'].mean()-stddev, my_dataset['Zr'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$')
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

q1 = np.percentile(my_dataset.Zr, 25, method ='midpoint') 
q3 = np.percentile(my_dataset.Zr, 75, method ='midpoint')
iqr = q3 - q1 # Interquaritle range (IQR)  

print ('-------')
print ('Interquaritle range (IQR): {0:.0f} [ppm]'.format(iqr))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist') 
ax.axvline(q1, color='purple', label='Q1', linewidth=2)
ax.axvline(q3, color='green', label='Q3', linewidth=2)
ax.axvspan(q1, q3, alpha=0.1, color='orange', label='Interquaritle range (IQR)')
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()

'''
Output:
-------
Interquaritle range (IQR): 164 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

a_mean = my_dataset.Zr.mean()

median = my_dataset.Zr.median()

hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] +  bin_edges[hist.argmax()+1])/2

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins=20, density=True, edgecolor='k', label="Measurements Hist") 
ax.axvline(modal_value, color='orange', label='Modal Value', linewidth=2)
ax.axvline(median, color='purple',  label='Median Value', linewidth=2)
ax.axvline(a_mean, color='green',  label='Arithmetic mean', linewidth=2)
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()

```

```python
import numpy as np
from scipy.stats import skew

a_mean = my_dataset.Zr.mean()
median = my_dataset.Zr.median()
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
standard_deviation = my_dataset['Zr'].std()

a1 = (a_mean - modal_value) / standard_deviation 
a2 = 3 * (a_mean - median) / standard_deviation
g1 = skew(my_dataset['Zr'])

print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```

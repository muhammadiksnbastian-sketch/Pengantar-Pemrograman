```python
count_0_to_4 = 5  # Variabel untuk menentukan jumlah iterasi dari 0 hingga 4

print('a sequence from 0 to 4')
for index_0_to_4 in range(count_0_to_4):
    print(index_0_to_4)  # Menampilkan setiap angka dari 0 hingga 4

print('----------------------')    
start_3_to_6 = 3  # Variabel untuk menentukan awal range dari 3
end_3_to_6 = 7    # Variabel untuk menentukan akhir range hingga 6

print('a sequence from 3 to 6')
for index_3_to_6 in range(start_3_to_6, end_3_to_6):
    print(index_3_to_6)  # Menampilkan setiap angka dari 3 hingga 6

print('----------------------')    
start_5_to_15 = 5  # Variabel untuk menentukan awal range dari 5
end_5_to_15 = 16   # Variabel untuk menentukan akhir range hingga 15
step_5_to_15 = 3   # Variabel untuk menentukan langkah setiap iterasi

print('a sequence from 5 to 15 with a step of 3')
for index_5_to_15 in range(start_5_to_15, end_5_to_15, step_5_to_15):
    print(index_5_to_15)  # Menampilkan setiap angka dari 5 hingga 15 dengan step 3
```

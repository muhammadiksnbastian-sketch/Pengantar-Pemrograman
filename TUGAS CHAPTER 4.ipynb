```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


def my_line(x, m, q):
    y = m * x + q
    return y


my_dataset = pd.read_excel(r'C:\Users\Iksan\python_earth_science_book-main\code\chapter_04\Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_majors', engine='openpyxl')

my_dataset1 = my_dataset[my_dataset.Epoch == 'one']
my_dataset2 = my_dataset[my_dataset.Epoch == 'three']
my_dataset3 = my_dataset[my_dataset.Epoch == 'three-b']

x = np.linspace(52.5, 62, 100)
y = my_line(x, m=0.3, q=-10.3)

fig, ax = plt.subplots()

ax.scatter(my_dataset1.SiO2, my_dataset1.K2O, marker='o', color="#cbfc06", edgecolor='#000000', label=r'$1^{st}$ Epoch')
ax.scatter(my_dataset2.SiO2, my_dataset2.K2O, marker='o', color="#16bcef", edgecolor='#000000', label=r'$3^{rd}$ Epoch')
ax.scatter(my_dataset3.SiO2, my_dataset3.K2O, marker='o', color="#4af583", edgecolor='#000000', label=r'$3^{-b}$ Epoch')
ax.plot(x, y, color='#342a77')

ax.annotate(r'What is the 1$\sigma$ for this point?', xy=(47.6, 6.6), xytext=(47, 8.8), arrowprops=dict(arrowstyle="->", connectionstyle="arc3"), fontsize=10)


ax.text(52.4, 5.6, r'$ Na_2O = 0.3 \cdot SiO_2 -10.3$', dict(size=10, rotation=33))

ax.text(53.5, 5.1, r'$ \mu_{SiO_2} = \frac {a_{1}+a_{2}+\cdots +a_{n}}{n}$ = ' + '{:.1f} [wt.%]'.format(57.721),  dict(size=11.5))

ax.set_xlabel(r'SiO$_2$ [wt%]')
ax.set_ylabel(r'K$_2$O [wt%]')

ax.legend()
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_title("Relationship between SiO2 and K2O in Different Epochs")
ax.set_facecolor("#C8C4AF")
```

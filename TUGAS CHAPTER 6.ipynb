```python
import pandas as pd  # mengimpor pandas sebagai pd
import matplotlib.pyplot as plt  # mengimpor matplotlib untuk visualisasi
import seaborn as sns  # mengimpor seaborn untuk visualisasi statistik

my_dataset = pd.read_excel(r'C:\Users\ThinkPad X13 Yoga\Documents\CODING\python_earth_science_book-main\python_earth_science_book-main\code\chapter_05\Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca data dari file Excel

my_sub_dataset = my_dataset[['Ce','La','U','Sc']]  # mengambil subset kolom Ce, La, U, dan Sc

cov = my_sub_dataset.cov()  # menghitung matriks kovarians antar variabel
cor = my_sub_dataset.corr()  # menghitung matriks korelasi antar variabel

fig = plt.figure(figsize=(11,5))  # membuat figure dengan ukuran tertentu

ax1 = fig.add_subplot(1,2,1)  # membuat subplot pertama (1 baris, posisi 1)
ax1.set_title('Covariance Matrix')  # memberi judul pada subplot pertama
sns.heatmap(cov, annot=True, cmap='cividis', ax=ax1)  # menampilkan heatmap kovarians dengan anotasi nilai

ax2 = fig.add_subplot(1,2,2)  # membuat subplot kedua (posisi 2)
ax2.set_title('Correlation Matrix')  # memberi judul pada subplot kedua
sns.heatmap(cor, annot=True, vmin= -1, vmax=1,  cmap='coolwarm', ax=ax2)  # menampilkan heatmap korelasi dengan skala -1 sampai 1

fig.tight_layout()  # mengatur layout agar tidak saling bertumpuk
```

```python
import pandas as pd  # mengimpor pandas sebagai pd
import scipy.stats as st  # mengimpor modul statistik dari scipy
import numpy as np  # mengimpor numpy untuk operasi numerik
import matplotlib.pyplot as plt  # mengimpor matplotlib untuk visualisasi

my_dataset = pd.read_excel(r'C:\Users\ThinkPad X13 Yoga\Documents\CODING\python_earth_science_book-main\python_earth_science_book-main\code\chapter_05\Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')  # membaca data dari file Excel

fig = plt.figure()  # membuat objek figure
ax1= fig.add_subplot(2,1,1)  # membuat subplot pertama (2 baris, posisi 1)
ax1.scatter(my_dataset.La, my_dataset.Ce, marker='o', edgecolor='k', color='#c7ddf4', label='CFC recent Activity')  # plot sebar La terhadap Ce
b1, b0, rho_value, p_value, std_err = st.linregress(my_dataset.La, my_dataset.Ce)  # menghitung parameter regresi linear
x = np.linspace(my_dataset.La.min(),my_dataset.La.max())  # membuat nilai x untuk garis regresi
y = b0 + b1*x  # menghitung nilai y berdasarkan persamaan garis regresi
ax1.plot(x, y, linewidth=1, color='#ff464a', linestyle='--', label=r"fit param.: $\beta_0$ = " + '{:.1f}'.format(b0) + r" - $\beta_1$ = "  + '{:.1f}'.format(b1) + r" - $r_{xy}^{2}$ = " + '{:.2f}'.format(rho_value**2))  # menambahkan garis regresi dan informasi parameter
ax1.set_xlabel('La [ppm]')  # label sumbu X
ax1.set_ylabel('Ce [ppm]')  # label sumbu Y
ax1.legend(loc= 'upper left')  # menampilkan legenda di kiri atas

ax2 = fig.add_subplot(2,1,2)  # membuat subplot kedua (posisi 2)
ax2.scatter(my_dataset.Sc, my_dataset.U, marker='o', edgecolor='k', color='#c7ddf4', label='CFC recent Activity')  # plot sebar Sc terhadap U
b1, b0, rho_value, p_value, std_err = st.linregress(my_dataset.Sc, my_dataset.U)  # menghitung parameter regresi linear
x = np.linspace(my_dataset.Sc.min(),my_dataset.Sc.max())  # membuat nilai x untuk garis regresi
y = b0 + b1*x  # menghitung nilai y berdasarkan persamaan garis regresi
ax2.plot(x, y, linewidth=1, color='#ff464a', linestyle='--', label=r"fit param.: $\beta_0$ = " + '{:.1f}'.format(b0) + r" - $\beta_1$ = "  + '{:.1f}'.format(b1) + r" - $r_{xy}^{2}$ = " + '{:.2f}'.format(rho_value**2))  # menambahkan garis regresi dan parameter
ax2.set_xlabel('Sc [ppm]')  # label sumbu X
ax2.set_ylabel('U [ppm]')  # label sumbu Y
ax2.legend(loc= 'upper left')  # menampilkan legenda di kiri atas
```

```python
import numpy as np  # mengimpor numpy untuk operasi numerik
import matplotlib.pyplot as plt  # mengimpor matplotlib untuk visualisasi

x = np.arange(1,6)  # membuat array x dari 1 sampai 5
y = np.array([0,1,2,9,9])  # membuat data y sebagai pasangan nilai x

fig, ax = plt.subplots()  # membuat objek figure dan axes
ax.scatter(x, y, marker = 'o', s = 100, color = '#c7ddf4', edgecolor = 'k')  # membuat scatter plot dari data x dan y

orders = np.array([2,3,4])  # menentukan orde polinomial yang akan digunakan
colors =['#ff464a','#342a77','#4881e9']  # menentukan warna untuk tiap kurva
linestiles = ['-','--','-.']  # menentukan jenis garis untuk tiap kurva

for order, color, linestile in zip(orders, colors, linestiles):   # melakukan iterasi untuk tiap kombinasi parameter
    betas = np.polyfit(x, y, order)  # menghitung koefisien polinomial sesuai orde
    func = np.poly1d(betas)  # membentuk fungsi polinomial dari koefisien
    x1 = np.linspace(0.5,5.5, 1000)  # membuat titik x baru untuk kurva halus
    y1 = func(x1)  # menghitung nilai y berdasarkan fungsi polinomial
    ax.plot(x1, y1, color=color, linestyle=linestile, label="Linear model of order " + str(order))  # menampilkan kurva hasil fitting

ax.legend()    # menampilkan legenda
ax.set_xlabel('A quantity relevant in geology\n(e.g., time)')  # memberi label sumbu X dengan baris baru
ax.set_ylabel('A quantity relevant in geology\n(e.g., spring flow rate)')  # memberi label sumbu Y dengan baris baru
fig.tight_layout()  # mengatur layout agar rapi
```

```python
import numpy as np  # mengimpor numpy untuk operasi numerik
import matplotlib.pyplot as plt  # mengimpor matplotlib untuk visualisasi
from scipy.optimize import curve_fit  # mengimpor fungsi untuk fitting kurva non-linear

def func(r, r0, D0, E):  # mendefinisikan fungsi model difusi berbasis radius ion
    R=8.314462618  # konstanta gas universal
    scale = 1e-21 # faktor skala untuk konversi satuan
    T = 800 + 273.15  # temperatur dalam Kelvin
    Na=6.02e23  # bilangan Avogadro
    return D0*np.exp((-4*np.pi*E*Na*((r0/2)*(r-r0)**2+(1/3)*(r-r0)**3)*scale)/(R*T))  # persamaan eksponensial model

def add_elements(ax):  # fungsi untuk menambahkan label elemen pada plot
    # to plot the name of the elements on the diagram
    names = ['La', 'Ce', 'Nd', 'Sm', 'Eu', 'Gd', 'Dy', 'Er', 'Yb', 'Lu', 'Y', 'Sc']  # nama elemen
    annotate_xs = np.array([1.172 + 0.01, 1.15 + 0.01, 1.123 + 0.01, 1.098 - 0.031, 1.087 - 0.028, 1.078 - 0.04, 1.052 + 0.005, 1.03 + 0.02, 1.008 + 0.01, 1.001 - 0.015, 1.04 -0.02, 0.885 - 0.03])  # posisi X label
    annotate_ys = np.array([0.468 + 0.1, 1.050 + 0.2, 10.305 + 3, 31.283 - 13, 45.634 -17, 74.633- 30, 229.279 + 80, 485.500, 583.828 +200, 460.404 -220, 172.844 -70, 141.630])  # posisi Y label
    
    for name, annotate_x, annotate_y in zip(names, annotate_xs, annotate_ys):  # iterasi tiap elemen
        ax.annotate(name, (annotate_x, annotate_y))  # menambahkan teks label pada plot

Di = np.array([0.468, 1.050, 10.305, 31.283, 45.634, 74.633, 229.279, 485.500, 583.828, 460.404, 172.844, 141.630])  # data nilai D_i
I_r = np.array([1.172, 1.15, 1.123, 1.098, 1.087, 1.078, 1.052, 1.03, 1.008, 1.001, 1.04, 0.885])  # data radius ion

fig = plt.figure(figsize=(9,5))  # membuat figure dengan ukuran tertentu

# Trust Region Reflective algorithm
ax1 = fig.add_subplot(1,2,1)  # subplot pertama (1 baris, posisi 1)
ax1.set_title("Trust Region Reflective algorithm")  # memberi judul subplot
ax1.scatter(I_r, Di, s=80, color='#c7ddf4', edgecolors='k', label='4 GPa - 1073 K, Kessel et al., 2005')  # plot data

popt1, pcov1 = curve_fit(func, I_r, Di, method='trf', bounds=([0.8,0,0],[1.3,1000,1000]))  # fitting kurva dengan metode TRF dan batas parameter

x1 = np.linspace(0.85,1.2,1000)  # membuat rentang nilai x untuk kurva fitting
y1 = func(x1,popt1[0],popt1[1], popt1[2])  # menghitung nilai y dari hasil fitting
ax1.plot(x1,y1, color='#ff464a', linewidth=2, linestyle ='--', label=r'$r_0$ = ' + '{:.3f}'.format(popt1[0]) + r', $D_0$ = ' + '{:.0f}'.format(popt1[1]) + ', E = ' + '{:.0f}'.format(popt1[2]))  # menampilkan kurva fitting dan parameter
add_elements(ax = ax1)  # menambahkan label elemen
ax1.set_yscale('log')  # mengatur skala logaritmik pada sumbu Y
ax1.set_xlabel(r'Ionic Radius ($\AA$)')  # label sumbu X
ax1.set_ylabel(r'$D_i$')  # label sumbu Y
ax1.set_ylim(0.005,3000)  # mengatur batas sumbu Y
ax1.legend()  # menampilkan legenda

# Levenberg-Marquardt algorithm
ax2 = fig.add_subplot(1,2,2)  # subplot kedua (posisi 2)
ax2.set_title("Levenberg-Marquardt algorithm")  # memberi judul subplot
ax2.scatter(I_r, Di, s=80, color='#c7ddf4', edgecolors='k', label='4 GPa - 1073 K, Kessel et al., 2005')  # plot data

popt2, pcov2 = curve_fit(func, I_r, Di, method='lm', p0=(1.1,100,100))  # fitting kurva dengan metode LM dan nilai awal parameter

x2 = np.linspace(0.85,1.2,1000)  # membuat rentang nilai x untuk kurva fitting
y2 = func(x2,popt2[0],popt2[1], popt2[2])  # menghitung nilai y dari hasil fitting
ax2.plot(x2,y2, color='#4881e9', linewidth=2, linestyle ='--', label=r'$r_0$ = ' + '{:.3f}'.format(popt2[0]) + r', $D_0$ = ' + '{:.0f}'.format(popt2[1]) + ', E = ' + '{:.0f}'.format(popt2[2]))  # menampilkan kurva fitting dan parameter
add_elements(ax = ax2)  # menambahkan label elemen
ax2.set_yscale('log')  # mengatur skala logaritmik pada sumbu Y
ax2.set_xlabel(r'Ionic Radius ($\AA$)')  # label sumbu X
ax2.set_ylabel(r'$D_i$')  # label sumbu Y
ax2.set_ylim(0.005,3000)  # mengatur batas sumbu Y
ax2.legend()  # menampilkan legenda

fig.tight_layout()  # merapikan layout agar tidak bertumpuk
```
